###  Bar Charts 

Wilderness Areas:

The bar charts for each wilderness area are very informatve and show that not all species grow in each wilderness area. This inforamtion will help the machine learning algorithm pair down the data and should make it much more accurate

```python
for i in col_names[11:15]:
    total = np.array(df.groupby(['Forest_Cover_Type'])['Slope'].count().tolist())
    x=[1,2,3,4,5,6,7]
    y = df.groupby(['Forest_Cover_Type'])[i].sum().tolist()
    y= y/total
    print(i)
    plt.bar(x,y)
    plt.show()
```

Soil Type:

Shown below are 4 examples of the soil type bar charts. They, like the wilderness areas, show that not all species will grow in each soil type. These will also be good to start from with our model and move out from there. Some soils will have all species growing on them while others only have 1 or 2. 

```python
for i in soil_type_code[4:8]:
    total = np.array(df.groupby(['Forest_Cover_Type'])['Slope'].count().tolist())
    x=[1,2,3,4,5,6,7]
    y = df.groupby(['Forest_Cover_Type'])[i].sum().tolist()
    y= y/total
    print(i)
    plt.bar(x,y)
    plt.show()
```
